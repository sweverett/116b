## 14.10.10 (HW 4 Problem 2)

Here's my solution for the second homework problem you turned in last week. First take the suggested map $w(z)=\sin(z)$ and compute $u(x,y)$ and $v(xy)$:

\begin{align*}
\sin(z)&=\sin(x+iy)\\
&=\sin(x)\cos(iy)+\cos(x)\sin(iy)\\
&=\sin(x)\cosh(y)+i\cos(x)\sinh(y)
\end{align*}

where I used a common trig identity (there are other ways to solve this using Euler's identity!). This means that

\begin{align*}
u(x,y)&=\sin(x)\cosh(y)\\
v(x,y)&=\cos(x)\sinh(y)
\end{align*}

Now the hint suggests to map the $u$-axis to the boundary of the tube. This is admittedly a bit tricky and takes some practice to know what to do. Note that on the $u$-axis, $v(x,y)=0$. Then

$$0=\cos(x)\sinh(y)$$

can only be true if $x=\pm\frac{\pi}{2}$, or if $\sin(y)=0\implies y=0$. Let's plot these lines below:

```python
pi = np.pi
M = 6 # bounds
N = 500 # number of points
x = np.linspace(-M,M,N)
y = np.linspace(-M,M,N)
```

```python
plt.subplot(1,2,1)
# W-Plane for V(x,y)=0
plt.plot([-M,M],[0,0],'k--') # u-axis
plt.xlabel('u')
plt.ylabel('v')
plt.title('W-Plane')

plt.subplot(1,2,2)
# Z-Plane for V(x,y)=0
plt.plot([-pi/2,-pi/2],[0,M],'k--') # x=-pi/2
plt.plot([-pi/2,pi/2],[0,0],'k--') # y=0
plt.plot([pi/2,pi/2],[0,M],'k--') # x=pi/2
plt.xlabel('x')
plt.ylabel('y')
plt.title('Z-plane')
plt.ylim([-1,M])
plt.xlim([-pi,pi])

plt.gcf().set_size_inches(10,5)
```

We've recovered our tube! Notice that the $u$-axis can also be thought of a line of constant $v(x,y)$ - namely $v=0$. What happens if we change the value of $v(x,y)=$constant? Let's try! First solve for $y(x)$ in terms of the constant $v$:

$$v(x,y)=v_c=\cos(x)\sinh(y)\implies y(x)=\sinh^{-1}\left(\frac{v_c}{\cos(x)}\right)$$

Now plot over a region of $x$ for lots of values of $v_c$:

```python
## Set up previous plot
plt.subplot(1,2,1)
# W-Plane for V(x,y)=0
plt.plot([-M,M],[0,0],'k--') # u-axis
plt.xlabel('u')
plt.ylabel('v')
plt.title('W-Plane')
plt.ylim([-1,M])

plt.subplot(1,2,2)
# Z-Plane for V(x,y)=0
plt.plot([-pi/2,-pi/2],[0,M],'k--') # x=-pi/2
plt.plot([-pi/2,pi/2],[0,0],'k--') # y=0
plt.plot([pi/2,pi/2],[0,M],'k--') # x=pi/2
plt.xlabel('x')
plt.ylabel('y')
plt.title('Z-plane')
plt.ylim([-1,M])

## plot contours

V = np.linspace(.1,2*pi,8)

for v in V:
    plt.subplot(1,2,1)
    plt.plot([-M,M],[v,v],'--')
    plt.subplot(1,2,2)
    y = np.arcsinh(v/np.cos(x))
    plt.plot(x,y)

plt.gcf().set_size_inches(10,5)
```

Wait - what's going on? This doesn't look right! A lot of you drew contours that looked like this because it's what Wolfram Alpha gave you (and me!). There are vertical lines and copies of what we care about because of domain issues with the inverse $\sinh$ function (this is why we introduce branch cuts in complex analysis!). To deal with this, you restrict the region you care about to just inside the tube - which we already decided was $-\pi/2\leq x\leq\pi/2$.

```python
#Fix domain of x
x = np.linspace(-pi/2,pi/2,N)

## Set up previous plot
plt.subplot(1,2,1)
# W-Plane for V(x,y)=0
plt.plot([-M,M],[0,0],'k--') # u-axis
plt.xlabel('u')
plt.ylabel('v')
plt.title('W-Plane')
plt.ylim([-1,M])

plt.subplot(1,2,2)
# Z-Plane for V(x,y)=0
plt.plot([-pi/2,-pi/2],[0,M],'k--') # x=-pi/2
plt.plot([-pi/2,pi/2],[0,0],'k--') # y=0
plt.plot([pi/2,pi/2],[0,M],'k--') # x=pi/2
plt.xlabel('x')
plt.ylabel('y')
plt.title('Z-plane')
plt.ylim([-1,M])
plt.xlim([-pi,pi])

## plot contours

V = np.linspace(.1,2*pi,8)

for v in V:
    plt.subplot(1,2,1)
    plt.plot([-M,M],[v,v],'--')
    plt.subplot(1,2,2)
    y = np.arcsinh(v/np.cos(x))
    plt.plot(x,y)

plt.gcf().set_size_inches(10,5)
```

 ## Problem 14.9.4 (modified)

I actually did this one during the review section on Friday, but I thought it would help to see it done fully in writing. The conformal map we are asked to plot is

$$w(z)=e^z$$

Let's solve for $u$ and $v$!

\begin{align*}
e^z&=e^{x+iy}\\
&=e^xe^{iy}\\
&=e^x\left[\cos(y)+i\sin(y)\right]\\
&=e^x\cos(y)+ie^x\sin(y)
\end{align*}

which means that

\begin{align*}
u(x,y)&=e^x\cos(y)\\
v(x,y)&=e^x\sin(y)
\end{align*}

Now in this problem, Boas actually asks you to hold $u$ and $v$ to be constant and plot the map in the $Z$-plane. However this is harder, so let's start with the opposite (and midterm-type) problem first.

What happens when we hold $x$ and $y$ constant? First look at the line $y=0$. Then we get

\begin{align*}
u(x,y)&=e^x\cos(0)=e^x\\
v(x,y)&=e^x\sin(0)=0
\end{align*}

What do we do with this information? Plot over a bunch of $x$, just like in the previous problem!

```python
xx = np.linspace(-M,M,N)
```

```python
plt.subplot(1,2,1)
# W-Plane for U(x,y)=e^x
u = np.exp(xx)
v = 0*xx
plt.plot(u,v)
plt.xlabel('u')
plt.ylabel('v')
plt.title('W-Plane')
plt.xlim(-M,M)
plt.ylim(-M,M)

plt.subplot(1,2,2)
# Z-Plane for y=0
plt.plot([-M,M],[0,0]) # x=-pi/2
plt.xlabel('x')
plt.ylabel('y')
plt.title('Z-plane')

plt.gcf().set_size_inches(10,5)
```

Well that wasn't very exciting. What if we let $y$ equal constants besides 0? Notice that we're essentially going through the same steps as the previous (and harder) problem!

```python
Y = np.linspace(0,2*pi,11)

for y in Y:
    plt.subplot(1,2,1)
    # W-plane for U(x,y)=e^xcos(y)
    u = np.exp(xx)*np.cos(y)
    v = np.exp(xx)*np.sin(y)
    plt.plot(u,v)
    plt.xlabel('u')
    plt.ylabel('v')
    plt.title('W-Plane')
    plt.xlim(-M,M)
    plt.ylim(-M,M)
    
    plt.subplot(1,2,2)
    # Z-Plane for y=constant
    plt.plot([-M,M],[y,y])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Z-plane')

plt.gcf().set_size_inches(10,5)
```

That's more interesting! What we have found is that lines of constant $y$ are mapped to rays in the $W$-plane, with $y$ giving the value of $\theta$ (which you can see by looking at $u$ and $v$).

But we're not done yet - we need to look at lines of constant $x$. This is done in an identical way, by plugging in constant values of x into $u(x,y)$ and $v(x,y)$ and plotting over a bunch of values of $y$:

```python
xx = np.linspace(-M,M,N)
yy = np.linspace(-M,M,N)

X = np.linspace(.1,3,8)
Y = np.linspace(0,2*pi,11)

for x in X:
    plt.subplot(1,2,1)
    # W-plane
    u = np.exp(x)*np.cos(yy)
    v = np.exp(x)*np.sin(yy)
    plt.plot(u,v,'k')
    plt.xlabel('u')
    plt.ylabel('v')
    plt.title('W-Plane')
    plt.xlim(-M,M)
    plt.ylim(-M,M)
    
    plt.subplot(1,2,2)
    # Z-Plane
    plt.plot([x,x],[-M,M],'k')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Z-plane')

for y in Y:
    plt.subplot(1,2,1)
    # W-plane
    u = np.exp(xx)*np.cos(y)
    v = np.exp(xx)*np.sin(y)
    plt.plot(u,v,'b')
    plt.xlabel('u')
    plt.ylabel('v')
    plt.title('W-Plane')
    plt.xlim(-M,M)
    plt.ylim(-M,M)
    
    plt.subplot(1,2,2)
    # Z-Plane
    plt.plot([-M,M],[y,y],'b')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Z-plane')
    plt.ylim([-0,6])
    plt.xlim([0,3])

plt.gcf().set_size_inches(10,5)
```

So we find that lines of constant $y$ (blue) map to rays of angle $\theta=y$ in the $W$-plane, while lines of constant x (black) map to circles of radius $e^x$. Not so bad! Also note that we didn't need a computer to solve this one. It can be done purely by looking at the $u$ and $v$ functions we found and asking what the limits of the functions are as you vary $x$ and $y$.

In case you're curious what the actual problem Boas assigned looks like, here it is: (ignore some sloppy coding)

```python
M = 5
N = 1000

xx = np.linspace(-5,6,N)

U = np.linspace(-2*pi,2*pi,30)
V = np.linspace(-2*pi,2*pi,30)

for u in U:
    plt.subplot(1,2,1)
    # W-plane
    plt.plot([u,u],[-M,M])
    plt.xlabel('u')
    plt.ylabel('v')
    plt.title('W-Plane')
    plt.xlim(-M,M)
    plt.ylim(-M,M)
    
    plt.subplot(1,2,2)
    # Z-Plane
    y = np.arccos(u/np.exp(xx))
    plt.plot(xx,y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Z-plane')

for v in V:
    plt.subplot(1,2,1)
    # W-plane
    plt.plot([-M,M],[v,v])
    plt.xlabel('u')
    plt.ylabel('v')
    plt.title('W-Plane')
    plt.xlim(-M,M)
    plt.ylim(-M,M)
    
    plt.subplot(1,2,2)
    # Z-Plane
    y = np.arcsin(v/np.exp(xx))
    plt.plot(xx,y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Z-plane')
    plt.ylim(0,pi/2)

plt.gcf().set_size_inches(10,5)
```

I don't think that problem is nearly as interesting! Note that the $Z$-plane doesn't look conformal (right angles) because of the scaling of the plot. If we look at a rescaled region we see that

```python
M = 5
N = 1000

xx = np.linspace(0,3,N)

U = np.linspace(-2*pi,2*pi,60)
V = np.linspace(-2*pi,2*pi,60)

for u in U:    
    y = np.arccos(u/np.exp(xx))
    plt.plot(xx,y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Z-plane')

for v in V:
    y = np.arcsin(v/np.exp(xx))
    plt.plot(xx,y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Z-plane')

plt.ylim(0,pi/2)
plt.xlim(0,3)
plt.gcf().set_size_inches(10,5)
```
